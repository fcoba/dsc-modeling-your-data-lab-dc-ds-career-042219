```python
subs = [(' ', '_'),('.',''),("'",""),('™', ''), ('®',''), ('+','plus'), ('½','half'), ('-','_') ]
def col_formatting(col):
    for old, new in subs:
        col = col.replace(old,new)
    return col
```
